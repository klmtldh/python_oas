```python
import pandas as pd
import matplotlib.pyplot as plt
from pylab import mpl
df = pd.DataFrame( [ [4,7,10], [5,8,11], [6,9,12] ], index=[1,2,3], columns=['a','b','c'] )
ax = df.plot()
fig = ax.get_figure()
fig.savefig(r'D:\JGY\600-Data\003-out输出文件\测试\fig.png')
plt.clf()

label_dis = df.a.value_counts() 
ax = label_dis.plot(title='label distribution', kind='barh', figsize=(18, 12),table=True) 
fig = ax.get_figure() 
fig.savefig(r'D:\JGY\600-Data\003-out输出文件\测试\label_distribution.png')
plt.clf()
```

```python
# import pandas as pd

speed = [0.1, 17.5, 40, 48, 52, 69, 88]
lifespan = [2, 8, 70, 1.5, 25, 12, 28]
index = ['snail', 'pig', 'elephant', 'rabbit', 'giraffe', 'coyote', 'horse']
df = pd.DataFrame({'speed': speed,'lifespan': lifespan}, index=index)
ax = df.plot.bar(rot=45)
plt.show()
plt.clf()
ax = df.plot.bar(stacked=True)
plt.show()
plt.clf()
axes = df.plot.bar(rot=0, subplots=True)
axes[1].legend(loc=2)  # doctest: +SKIP
plt.show()

axes = df.plot.bar(rot=0, subplots=True, color={"speed": "red", "lifespan": "green"}
)
axes[1].legend(loc=2)  # doctest: +SKIP
plt.show()

ax = df.plot.bar(y='speed', rot=0)
plt.show()
ax = df.plot.bar(x='lifespan', rot=0)
plt.show()
```

```python
import pandas as pd
def pandas_rat(df):
    df_table=df.pivot_table(index=['责任科室','责任人员'],
                       columns=['是否完成'], 
                       values=['业务领域'], 
                       aggfunc=len,
                       fill_value=0,
                       margins=True
                      )
    df_table.reset_index(inplace=True)
    df_table.columns=['责任科室','责任人员', '否', '是', '合计']
    df_table.loc[len(df_table)-1,'责任科室']= '安监部'
    df_table['完成率'] = df_table['是']/df_table['合计']
    df_table['完成率'] = df_table['完成率'].apply(lambda x: '%.2f%%' % (x*100))
    df_table['科室否'] =df_table.groupby('责任科室')['否'].transform('sum')
    df_table['科室是'] = df_table.groupby('责任科室')['是'].transform('sum')
    df_table['科室合计'] = df_table.groupby('责任科室')['合计'].transform('sum')
    df_table['科室完成率'] = df_table['科室是']/df_table['科室合计']
    df_table['科室完成率'] = df_table['科室完成率'].apply(lambda x: '%.2f%%' % (x*100))
    
    return df_table
if __name__ == '__main__':
    name = r'D:\JGY\600-Data\002-in输入文件\02-work工作\03-document工作文档\002-计划\昆明供电局安全监管重点工作计划.xlsx'
    ef = pd.ExcelFile(name, engine='openpyxl')
    df = ef.parse('昆明供电局安全监管重点工作计划')
    df_rat=pandas_rat(df)
    print(df_rat)
```
